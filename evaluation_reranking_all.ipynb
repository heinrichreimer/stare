```python
from pandas import DataFrame, read_csv
```

```python
df_original: DataFrame = read_csv("data/metrics-reranking-all-original.csv")
df_original
```

```python
df_gt: DataFrame = read_csv("data/metrics-reranking-all-ground-truth.csv")
df_gt
```

```python
df_gpt: DataFrame = read_csv("data/metrics-reranking-all-gpt3.csv")
df_gpt
```

```python
for k, df in {"original": df_original, "ground truth": df_gt, "GPT-3": df_gpt}.items():
    df = df[df["name"].str.endswith(" + stance-first@5 + alternating-stance@5")]
    for cols in [
        ["nDCG@5 rel.", "nDCG@5 qual."],
        ["alpha_nDCG@5 rel.", "alpha_nDCG@5 qual."],
    ]:
        df2 = df.copy()
        for col in cols:
            df2 = df2[df2[f"{col} test"].notna()]
            if len(df2) > 0:
                df2 = df2[df2[f"{col} test"].str.contains("1")]
        print(k, cols, len(df2) / 26)
```

```python
df_f1: DataFrame = read_csv("data/metrics-reranking-all-f1.csv")
df_f1
```

```python
eval = {}
for cols in [
    ("nDCG@5 rel.", "nDCG@5 qual."),
    ("alpha_nDCG@5 rel.", "alpha_nDCG@5 qual."),
]:
    eval[cols] = {}
    df1 = df_f1.copy()
    for f1 in [1, 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3]:
        df2 = df1.copy()
        f1_str = f"(F1<={f1:.2f})" if f1 < 1 else ""
        gt_string = f" + ground-truth{f1_str} + "
        df2 = df2[df2["name"].str.contains(gt_string, regex=False)]
        for col in cols:
            df2 = df2[df2[f"{col} test"].notna()]
            if len(df2) > 0:
                df2 = df2[df2[f"{col} test"].str.contains("1")]
        eval[cols][f1] = len(df2)
df_eval_f1 = DataFrame([
    {
        "cols": ", ".join(cols),
        "f1": f1,
        "n": n,
    }
    for cols, evals in eval.items()
    for f1, n in evals.items()
])
df_eval_f1
```
